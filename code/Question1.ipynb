```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import pylab as plt
import holidays
import pmdarima
from scipy import stats

path1 =  r"D:\MCM\data sigma.xlsx"#path1是短期的文件路径
df= pd.read_excel(path1,index_col='Date', parse_dates=True)

ind = df.index

df['percentage'] = 0
total_score = df['Number of  reported results']
hard_score = df['Number in hard mode']
list1 = []
for i in range(0,len(df)):
    list1.append(hard_score[i]/total_score[i])
df['percentage'] = list1

y = df['percentage']
x = df['holiday/workday']

x = x.replace('holiday', 1)
x = x.replace('workday', 0)

start_date = '2023-01-01'
end_date = '2023-3-01'
dates = pd.date_range('20230101',periods = 60)
us_holidays = holidays.US()
df2 = pd.DataFrame({"A":dates,"B":0})
d = df2['A']
h = df2['B']
df2.set_index("A",inplace = True)

for i in range(0,len(dates)):
    weekday = d[i].weekday()
    if d[i] in us_holidays or weekday ==5 or weekday ==6:
        h[i] = 'holiday'
    else:
        h[i] = 'workday'

h = h.replace('holiday', 1)
h = h.replace('workday', 0)

# 预测未来 n 个时间点的数据，其中 exog 表示外部变量


model = sm.tsa.SARIMAX(endog = y, order=(4, 1, 2), exog = x)
result = model.fit()

n = len(dates)
pred = result.forecast(steps=n, exog=h, alpha=0.05)

parray = result.fittedvalues
predictions_ARIMA_diff = pd.DataFrame(result.fittedvalues,index=ind, copy=True)

list1 = []
for i in range(0,len(parray)):
    list1.append(y[i] - parray[i])

print(list1)
list1_var = np.var(list1)
print("*******************************************************************************")
print(list1_var)
plt.plot(ind[0:] ,list1)
plt.show()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(y.index, y.values, label='Actual')
plt.plot(predictions_ARIMA_diff,label="check")
ax.plot(d, pred, label='Predicted',color = 'grey')



```
